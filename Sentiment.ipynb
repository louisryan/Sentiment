In the past number of years, there has been a huge surge in social platforms such as Twitter and Facebook to the point where they are ubiquitous. As more and more companies use Social Media Marketing to promote their brand, it becomes increasingly important for them to evaluate the effectiveness of their campaigns. One such measure is Sentiment Analysis. Sentiment Analysis allows you to determine the polarity of the customer towards particular content or campaigns and allows you to adjust your strategy accordingly.

Performing Sentiment Analysis on Twitter or Facebook data is a more complex challenge than doing it for larger documents, primarily due to the shorthand version of english often used by people, alternate puncuation and grammatical errors. This includes entities such as hashtages, emoticons, slang and other Twitter specific lingo. 

The system developed below uses advanced statistical models (machine learning & NLP) *trained* on social data, meaning the detection can handle slang, common misspellings, emoticons, hashtags, etc.

## Data

In any supervised learning problem, the first task is to collect labelled training data to allow for this problem to be formulated as a classification problem. With common machine learning projects such as Churn detection, it's a relatively straight forward task to extract labelled training data as often times, you record customer interactions with a product/service and thus, can conclude if a customer has left your service by some period of inactivity. 

With the large range of topics discussed on Twitter, it would be very difficult to manually collect enough data to train a sentiment classifier for tweets. Our solution is to use [distant supervision](http://stats.stackexchange.com/questions/46685/distant-supervision-supervised-semi-supervised-or-both), in which our training data consists of tweets with emoticons. This approach was introduced by Read [10]. The emoticons serve as noisy labels. For example, :) in a tweet indicates that the tweet contains positive sentiment and :( indicates that the tweet contains negative sentiment. With the help of the Twitter API and a [previous project](https://github.com/louisryan/Twitter_Analysis), it was easy to extract large amounts of tweets with emoticons in them. This is a significant improvement over the many hours it may otherwise take to hand-label training data. We run classifiers trained on emoticon data against a test set of tweets (which may or may not have emoticons in them).

## Naive Bayes Algorithm

From wikipedia:

>A naive Bayes classifier is a simple probabilistic classifier based on applying Bayes' theorem with strong (naive) independence assumptions. 

To formalize the Naive Bayes classifier we use Bayes Rule. It states that for a document(d) and a class(c), we can apply the formula below:

$$\Pr(c|d)= \frac{\Pr(d|c)\Pr(c)} {\Pr(d)}$$

We can drop the demoninator as the Probability of a document will be a constant. The probability of the class is a straightforward calculation(How often does this class occur?). What's slightly more complicated is the calculation:

$$ \Pr(d|c) $$

To make this calculation possible, we'll make the simplifying assumption that the feature probabilities are independent given the class. For example, a fruit may be considered to be an apple if it is red, round, and about 3" in diameter. A naive Bayes classifier considers each of these features to contribute independently to the probability that this fruit is an apple, regardless of the presence or absence of the other features. We know this is a terrible simplification to make but it works!

We'll use the _MultinomialNB_ Naive Bayes implementation offered by sklearn as it asssumes the features to be occurence counts, which is how we'll be encoding the tweets as features. In practice, it also works well with TF-IDF vectors. The method below constructs a [TfidfVectorizer](http://en.wikipedia.org/wiki/Tf–idf) to convert the raw tweets into the TF-IDF feature values, which we then use to train our classifier.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB

def create_ngram_model():
    tfidf_ngrams = TfidfVectorizer()
    clf = MultinomialNB()
    pipeline = Pipeline([('vect', tfidf_ngrams), ('clf', clf)])
    return pipeline
```

The following method trains the Naive Bayes algorithm on cross-validated data using ShuffleSplit. For each fold, we keep track of the area under the Precision-Recall curve and the accuracy

```python
from sklearn.metrics import precision_recall_curve, auc
from sklearn.cross_validation import ShuffleSplit
from sklearn.metrics import precision_recall_curve, roc_curve, auc
from sklearn.cross_validation import ShuffleSplit

def train_model(clf_factory, X, Y, name="NB ngram", plot=False):
    
    cv = ShuffleSplit(n=len(X), n_iter=10, test_size=0.3, indices=True, random_state=0)

    train_errors, test_errors = [],[]
    scores, pr_scores = [],[]
    precisions, recalls, thresholds = [], [], []

    for train, test in cv:
        X_train, y_train = X[train], Y[train]
        X_test, y_test = X[test], Y[test]
        
        clf = clf_factory()
        clf.fit(X_train, y_train)

        train_score = clf.score(X_train, y_train)
        test_score = clf.score(X_test, y_test)

        train_errors.append(1 - train_score)
        test_errors.append(1 - test_score)

        scores.append(test_score)
        proba = clf.predict_proba(X_test)

        fpr, tpr, roc_thresholds = roc_curve(y_test, proba[:, 1])
        precision, recall, pr_thresholds = precision_recall_curve(y_test, proba[:, 1])

        pr_scores.append(auc(recall, precision))
        precisions.append(precision)
        recalls.append(recall)
        thresholds.append(pr_thresholds)

    scores_to_sort = pr_scores
    median = np.argsort(scores_to_sort)[len(scores_to_sort) / 2]

    if plot:
        plot_pr(pr_scores[median], name, "01", precisions[median], recalls[median], label=name)
        summary = (np.mean(scores), np.std(scores), np.mean(pr_scores), np.std(pr_scores))
        #print "%.3f\t%.3f\t%.3f\t%.3f\t" % summary

    return summary
```

This is a handy function to plot the Precision/Recall graph below:

```python
def plot_pr(auc_score, name, phase, precision, recall, label=None):
    
    from matplotlib import pylab
    pylab.clf()
    pylab.figure(num=None, figsize=(5, 4))
    pylab.grid(True)
    pylab.fill_between(recall, precision, alpha=0.5)
    pylab.plot(recall, precision, lw=1)
    pylab.xlim([0.0, 1.0])
    pylab.ylim([0.0, 1.0])
    pylab.xlabel('Recall')
    pylab.ylabel('Precision')
    pylab.title('P/R curve (AUC=%0.2f) / %s' % (auc_score, label))
    filename = name.replace(" ", "_")
```

```python
# The raw feature(text) and response variable(labels)
X = sentiment['SentimentText']
y = sentiment['Sentiment']

summary = train_model(create_ngram_model, X, y, name="pos vs neg", plot=True)
```

To evaluate the effectiveness of our classifier, let's step back a bit. Our classifier doesn't necessarily have to predice both positive and negative tweets perfectly. We can tune the classifier to be particularly good for one class, we can adapt feedback to the user. For example, if the business case for this tool is to isolate customers who are speaking negatively about our brand, then we care mostly about _True Negative_. What we want is to have a high success rate when we are predicting a post to be negative, precision:

$$Precision = \frac {TP} {TP + FP}$$

If instead our goal was to detect as many positive or negative tweets, we would be more interested in recall:

$$Recall = \frac {TP} {TP + FN}$$

```python
print('The classifier has an accuracy of %0.2f with a standard deviation of %0.4f' %(summary[0:2]))
print('The classifier has a P/R AUC of score of %0.2f with a standard deviation of %0.4f' %(summary[2:4]))
```

### Tuning Parameters

There are a couple of areas that could be tuned to improve the performance of the classifier: _TfidfVectorzer_ and _MultinomialNB_. Let's try a combination of different paramters for optimization:

TfidfVectorizer:

* Use different settings for NGrams: unigramgs(1,1), bigrams(1,2) and trigrams(1,3). 
* Play with min_df: 1 or 2.
* Test removal of stop words: stop_words='English' or stop_words='None'
* Experiment with whether to track word counts or simply word occurencess.

MultinomiaNB:

* Decide on the smoothing method to use.
* Add-one or Laplace smoothing: 1
* Lidstone smoothing: 0.01, 0.05, 0.1, 0.5

We can use scikit-learns GridSearchCV class, using our instance of the classifier along with a dictionary of parameters with their potential values. Since we want to evaluate the classifier with our current best one, we pass the ShuffleSplit instance using the CV parameter. 

Finally, we pass GridSearchCV a score function to base it's decision on which is the best combination of hyperparamters. One metric that combines both precision and recall is the F-measure metric:

$$F = \frac {2.precision.recll}{precision+recall}$$

The following code is training 3.2.2.2.2.2.6 = 1,152 parameter combinations - each being trained on 10 folds

```python
from sklearn.grid_search import GridSearchCV
from sklearn.metrics import f1_score

def grid_search_model(clf_factory, X, y):
    cv = ShuffleSplit(n=len(X), n_iter=10, test_size=0.3, indices=True, random_state=0)
    
    param_grid = dict(vect__ngram_range=[(1, 1), (1, 2)],
                      vect__min_df=[1, 2],
                      vect__stop_words=[None, "english"],
                      vect__smooth_idf=[False, True], 
                      vect__use_idf=[False, True],
                      vect__sublinear_tf=[False, True],
                      vect__binary=[False, True],
                      clf__alpha=[0, 0.01, 0.05, 0.1, 0.5, 1],
                      )
    
    grid_search = GridSearchCV(clf_factory(), param_grid=param_grid, cv=cv, scoring =f1_score, verbose=10, n_jobs=2)
    grid_search.fit(X, y)
    clf = grid_search.best_estimator_
    print clf
    return clf

clf_gs = grid_search_model(create_ngram_model, X, y)
```

Due to a bug in the pipelining process, the gridsearch was not successful. Otherwise, the process would be to exhaustively search the parameter space and once the process identifies the optimal parameters, manually create the pipeline with those parameters like below:

```python
def get_best_model():
    best_params = dict(vect__ngram_range=(1, 2),
                       vect__min_df=1,
                       vect__stop_words=None,
                       vect__smooth_idf=False,
                       vect__use_idf=False,
                       vect__sublinear_tf=True,
                       vect__binary=False,
                       clf__alpha=0.01,
                       )

    best_clf = create_ngram_model(best_params)
    return best_clf
```
